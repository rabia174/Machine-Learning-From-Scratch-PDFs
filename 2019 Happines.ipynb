```python
# For example, here's several helpful packages to load in 

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
data2019=pd.read_csv('C:/Users/pc/Desktop/2019.csv')
```

```python
f,ax = plt.subplots(figsize=(18, 18))
sns.heatmap(data2019.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
plt.show()
```

```python
x=data2019['Country or region'].values[0:5]
y=data2019['Freedom to make life choices'].values[0:5]

plt.plot(x,y)
```

```python
x=data2019['Overall rank'].values
y=data2019['GDP per capita'].values

plt.plot(x,y,color='r')
plt.xlabel('Country Rank')
plt.ylabel('GDP per capita')
plt.suptitle('The relation between the rank & GDP')
plt.show()
```

```python
x=data2019['Overall rank'].values
y=data2019['Social support'].values

plt.scatter(x,y,color='green')
plt.xlabel('Overall rank')
plt.ylabel('Social support')
plt.suptitle('The relation between the rank & Social support')
plt.show()
```

```python
fig, axes = plt.subplots(figsize=(10, 10),nrows=2, ncols=2)
data2019.set_index('Overall rank')
data2019['Social support'].plot(ax=axes[0,0],kind='line' ,color='green' ,title='Social Support' ,linewidth=1 , grid=True ,linestyle=':' )
data2019['Generosity'].plot(ax=axes[0,1],kind='line' ,color='red' ,title='Generosity' ,linewidth=1 , grid=True ,linestyle=':' )
data2019['Healthy life expectancy'].plot(ax=axes[1,0],kind='line' ,color='yellow' ,title='Healthy life expectancy' ,linewidth=1 , grid=True ,linestyle=':' )
data2019['Perceptions of corruption'].plot(ax=axes[1,1],kind='line' ,color='blue' ,title='Perceptions of corruption' ,linewidth=1 , grid=True ,linestyle=':' )
```

```python
index=range(len(data2019['Country or region'].values[:10]))

plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(20, 10))
plt.yticks(index, data2019['Country or region'].values[:10])
ax.barh(data2019['Country or region'].values[:10], data2019['Generosity'].values[:10], color='yellow')
```
